```python
from matplotlib.pyplot import subplots
import numpy as np
import math as m
```

```python
def diff_toss(num_toss):
    x = np.random.uniform(0,1,size=num_toss)
    a = x[x >= 0.5].size
    b = x[x < 0.5].size
    return m.fabs(b-a)/num_toss
```

```python
x = np.arange(10,10000,10)
y = []
for i in x:
    y.append(diff_toss(i))
```

```python
fig,ax = subplots(figsize=(8,8))
ax.scatter(x,y,marker='o');
```
